Having identified the above anomalies, Feature Engineering is all about fixing them.

We need to seperate the date into day, month and year for ML use.

```python
# Fixing Order_Date Columns
data["Order_Date"].head()
```

```python
# Seeperating Order_Date Into Components, Convert Same To Int And Drop Original Parent Column
data["Order_day"]=data["Order_Date"].str.split("-").str[0]
data["Order_month"]=data["Order_Date"].str.split("-").str[1]
data["Order_day"]=data["Order_day"].astype(int)
data["Order_month"]=data["Order_month"].astype(int)
data.drop("Order_Date", axis=1, inplace=True)
```

```python
# Fixing Time_Orderd Columns

# Replace Non Dates Values With NaN:
non_time_value =['0.458333333','0.958333333','0.791666667','0.875','1','0.375','0.625','0.833333333','0.666666667', '0.75','0.416666667','0.916666667','0.5','0.708333333','0.541666667','0.583333333']
for i in non_time_value:
    data["Time_Orderd"]=data["Time_Orderd"].replace(i,np.NaN)
```

```python
data["Time_Orderd"].unique()
```

```python
# Split The Time Values Into Seperate Components AND Replace Missing Value
data["Time(hour)_order"]=data["Time_Orderd"].str.split(":").str[0]
data["Time(min)_order"]=data["Time_Orderd"].str.split(":").str[1]
data.drop('Time_Orderd', axis =1, inplace=True)

# data["Time(hour)_order"]= imp.fit_transform(data[["Time(hour)_order"]])
# data["Time(min)_order"]= imp.fit_transform(data[["Time(min)_order"]])
```

```python
# Fixing Time_Orderd Columns
data["Time_Order_picked"].unique()
```

```python
# Changing The Non-Time Values To NaN
top_unique=data["Time_Order_picked"].unique()

# Identify And Seperate Non-Time Values Into A List 
new_list=[i for i in top_unique if "." in i or i.isnumeric()]  
new_list
```

```python
for i in new_list:
    data["Time_Order_picked"]=data["Time_Order_picked"].replace(i,np.nan)
```

```python
data["Time_Order_picked"].unique()
```

```python
# Split The Time Values Into Seperate Components AND Replace Missing Value
data["Time(hour)_order_pk"]=data["Time_Order_picked"].str.split(":").str[0]
data["Time(min)_order_pk"]=data["Time_Order_picked"].str.split(":").str[1]
data.drop("Time_Order_picked", axis=1, inplace=True)
```
